```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
%matplotlib notebook
pg = "http://en.wikipedia.org/wiki/Delhi_Daredevils"
dfs_dd = pd.read_html(pg)
dfs_dd[8].to_csv('ex.csv')
df_win_loss = pd.read_csv('ex.csv')
df_win_loss
```

```python
df_win_loss_dd = df_win_loss.copy()
cols = ['Edition', '% win']
win_loss_dd = df_win_loss_dd[cols]
win_loss_dd # reading columns required
```

```python
win_loss_dd_cleaned = win_loss_dd[:12] # slicing the last row  
win_loss_dd_cleaned
```

```python
win_loss_dd_cleaned = win_loss_dd_cleaned.replace(to_replace = '50%', value = '50.00%')
win_loss_dd_cleaned
```

```python
for i in np.arange(0,12,1):
   
    win_loss_dd_cleaned['% win'][i] = win_loss_dd_cleaned['% win'][i][:-1]
    win_loss_dd_cleaned['Edition'][i] = win_loss_dd_cleaned['Edition'][i][3:]
    
win_loss_dd_cleaned
```

```python
edition_list = []
win_per_list = []
edition_list = win_loss_dd_cleaned['Edition']
win_per_list = win_loss_dd_cleaned['% win']
```

```python
win= np.array(win_per_list).tolist()
win
win_percent =[float(i) for i in win]
```

```python
y_pos = np.arange(len(edition_list))
y_pos
```

```python
plt.bar(y_pos,win_percent, label="Delhi Capitals Win %")
plt.xticks(y_pos, edition_list)
plt.ylabel('Win %')
plt.xlabel('IPL Edition')
plt.title('Delhi Capitals Win %')
plt.show()
```
